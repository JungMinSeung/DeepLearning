```python
# 2. 입력/출력 데이터 만들기
import numpy as np
X= np.array([[0,0],[0,1],[1,0],[1,1]])
y=np.array([[0],[0],[0],[1]])
```

```python
# 3. Keras 패키지 가져오기

from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD
```

```python
# 4. Perceptron 모델 생성 = 레고 머리만들기
model=Sequential()

model.add(Dense(units=1, input_dim=2, activation='sigmoid')) #fully connected layer , units는 출력 노드 갯수, input_dim은 input feature의 갯수

print(model.summary()) #params 는 feature 2개 + bias 1개 = 3개 ==> param # = (input 개수 + 1) x 노드개수
```

```python
# 5. Compile - Optimizer, Loss function 설정
sgd=SGD(lr=0.1)
model.compile(loss='binary_crossentropy',optimizer=sgd) # 레고 머리, 몸통, 다리 합치기
```

```python
# 6. 학습시키기

model.fit(X,y,batch_size=1, nb_epoch=500) 
# batch_size = 문제 몇개를 풀고 답을 확인 할 것인가? 한문제 풀고 확인하고 하면 느리지만 정확도 높음. 한번에 다풀고 정답 확인하면 빠른데 정확도 구림.
# epoch = interation. 데이터를 몇번을 반복해서 학습할 것인가.(문제집을 몇 번 풀것인가)
#loss=실제값-예측값
```

```python
# 7. 모델 테스트하기
test = np.array([[0,1]])
pred=model.predict(test)

print(pred)
```
